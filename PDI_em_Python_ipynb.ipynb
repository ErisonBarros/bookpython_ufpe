# Bibliotecas

```python
from spectral import imshow, matched_filter
import tifffile as tif
```

# leitura de Imagem

```python
!wget -P https://github.com/ErisonBarros/bookpython_ufpe/blob/master/Planta_topogr%C3%A1fica.tif
```

```python
img=tif.imread('/content/Planta_topográfica.tif')
```

# Visualização

```python
imshow(img,(2,3,1))
```

# Pixel Referencia 

```python
t=img[1,1]
t
```

# Calculando scores

```python
scores=matched_filter(img, t)
```

# Visulaizar resultado

```python
imshow(scores > 0.3)
```
